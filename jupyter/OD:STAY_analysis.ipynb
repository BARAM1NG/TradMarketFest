```python
import pandas as pd

def try_different_encodings(file):
    """ì—¬ëŸ¬ ì¸ì½”ë”©ì„ ì‹œë„í•˜ì—¬ CSV íŒŒì¼ì„ ì½ëŠ” í•¨ìˆ˜"""
    encodings = ['utf-8', 'cp949', 'euc-kr']

    for encoding in encodings:
        try:
            return pd.read_csv(file, encoding=encoding)
        except UnicodeDecodeError:
            continue
        except Exception as e:
            print(f"Error reading {file} with {encoding} encoding: {str(e)}")
            continue

    raise ValueError(f"Could not read {file} with any of the attempted encodings")

# YouTube ë°ì´í„° ë³‘í•©
def merge_youtube_data():
    youtube_files = [
        "ëŒ€êµ¬ ì¹˜ë§¥ í˜ìŠ¤í‹°ë²Œ_youtube_video_info_with_stats.csv",
        "ì„œë¬¸ì‹œì¥_youtube_video_info_with_stats.csv",
        "ì˜ˆì‚° ë§¥ì£¼ í˜ìŠ¤í‹°ë²Œ_youtube_video_info_with_stats.csv",
        "ì˜ˆì‚° ì‹œì¥_youtube_video_info_with_stats.csv"
    ]

    youtube_dfs = []

    for file in youtube_files:
        try:
            df = try_different_encodings(file)
            df['source_file'] = file.split('_')[0]
            youtube_dfs.append(df)
        except Exception as e:
            print(f"Error processing {file}: {str(e)}")

    if not youtube_dfs:
        raise ValueError("No YouTube data files were successfully read")

    youtube_merged = pd.concat(youtube_dfs, ignore_index=True)
    return youtube_merged

# ì™¸ì‹ ë°ì´í„° ë³‘í•©
def merge_dine_out_data():
    dine_out_files = [
        "dine_out_ëŒ€êµ¬ê´‘ì—­ì‹œ_ëŒ€ì‹ ë™_202308.csv",
        "dine_out_ì¶©ì²­ë‚¨ë„_ì˜ˆì‚°_202308.csv",
        "dine_out_ëŒ€êµ¬ê´‘ì—­ì‹œ_ëŒ€ì‹ ë™_202309.csv",
        "dine_out_ì¶©ì²­ë‚¨ë„_ì˜ˆì‚°_202309.csv"
    ]

    dine_out_dfs = []

    for file in dine_out_files:
        try:
            df = try_different_encodings(file)
            # íŒŒì¼ëª…ì—ì„œ ì •ë³´ ì¶”ì¶œ
            parts = file.split('_')
            region = parts[2]
            date = parts[-1].replace('.csv', '')

            df['region'] = region
            df['date'] = date
            dine_out_dfs.append(df)
        except Exception as e:
            print(f"Error processing {file}: {str(e)}")

    if not dine_out_dfs:
        raise ValueError("No dine-out data files were successfully read")

    dine_out_merged = pd.concat(dine_out_dfs, ignore_index=True)
    return dine_out_merged

# ë©”ì¸ ì‹¤í–‰ ë¶€ë¶„
def main():
    try:
        # YouTube ë°ì´í„° ë³‘í•©
        youtube_df = merge_youtube_data()
        print("YouTube ë°ì´í„° ë³‘í•© ì™„ë£Œ")
        print(f"ë³‘í•©ëœ YouTube ë°ì´í„° í¬ê¸°: {youtube_df.shape}")
        print("YouTube ë°ì´í„° ì»¬ëŸ¼:", youtube_df.columns.tolist())

        # ì™¸ì‹ ë°ì´í„° ë³‘í•©
        payment_df  = merge_dine_out_data()
        print("\nì™¸ì‹ ë°ì´í„° ë³‘í•© ì™„ë£Œ")
        print(f"ë³‘í•©ëœ ì™¸ì‹ ë°ì´í„° í¬ê¸°: {payment_df .shape}")
        print("ì™¸ì‹ ë°ì´í„° ì»¬ëŸ¼:", payment_df .columns.tolist())

        # ê²°ê³¼ ì €ì¥
        youtube_df.to_csv("merged_youtube_data.csv", index=False, encoding='utf-8-sig')
        payment_df .to_csv("merged_dine_out_data.csv", index=False, encoding='utf-8-sig')
        print("\në³‘í•©ëœ ë°ì´í„°ê°€ ì„±ê³µì ìœ¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

        return youtube_df, payment_df

    except Exception as e:
        print(f"Error during execution: {str(e)}")
        return None, None

if __name__ == "__main__":
    youtube_df, dine_out_df = main()
```

```python
import pandas as pd
import glob
from datetime import datetime
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ì„ íƒëœ ë„ì°©ì§€ ì½”ë“œ
dest_codes = [4481025000, 2711059500]  # ì˜ˆì‚°ì‹œì¥, ì„œë¬¸ì‹œì¥

# ì œì™¸í•  ì¶œë°œì§€ ì½”ë“œ
exclude_origin_codes = dest_codes

# í¬í•¨í•  ëª©ì  ì½”ë“œ
include_purpose = [3, 5]  # ì‡¼í•‘/ì—¬ê°€, ì—¬í–‰

# ë¶„ì„ ê¸°ê°„ ì„¤ì •
start_date = datetime.strptime('2023-09-01', '%Y-%m-%d')
end_date = datetime.strptime('2023-10-15', '%Y-%m-%d')

# íŒŒì¼ëª…ì—ì„œ ë‚ ì§œ ì¶”ì¶œ ë° ë‚ ì§œ ë²”ìœ„ë¡œ í•„í„°ë§í•˜ëŠ” í•¨ìˆ˜
def filter_files_by_date(file_list, start_date, end_date):
    filtered_files = []
    for file in file_list:
        filename = os.path.basename(file)
        date_str = filename.split('_')[1]  # 'YYYYMMDD'
        file_date = datetime.strptime(date_str, '%Y%m%d')
        if start_date <= file_date <= end_date:
            filtered_files.append(file)
    return filtered_files
```

```python
# OD ë°ì´í„° íŒŒì¼ ê²½ë¡œ (ì••ì¶• í•´ì œëœ íŒŒì¼)
od_csv_files = glob.glob('/content/data/od/*.csv')

# ì›í•˜ëŠ” ë‚ ì§œ ë²”ìœ„ì˜ OD íŒŒì¼ë§Œ ì„ íƒ
od_csv_files_filtered = filter_files_by_date(od_csv_files, start_date, end_date)

# OD ë°ì´í„°ë¥¼ í•„í„°ë§í•˜ëŠ” í•¨ìˆ˜
def filter_od_data(file):
    od_df = pd.read_csv(file, dtype={'origin_hdong_cd': 'int64', 'dest_hdong_cd': 'int64',
                                     'date': 'int64', 'od_cnts': 'int32',
                                     'dest_purpose': 'int8'})

    # ë‚ ì§œ í•„í„°ë§
    od_df['date'] = pd.to_datetime(od_df['date'], format='%Y%m%d')
    od_df = od_df[(od_df['date'] >= start_date) & (od_df['date'] <= end_date)]

    # í•„í„°ë§ ì¡°ê±´ ì ìš©
    filtered_od = od_df[
        (od_df['dest_hdong_cd'].isin(dest_codes)) &
        (~od_df['origin_hdong_cd'].isin(exclude_origin_codes)) &
        (od_df['dest_purpose'].isin(include_purpose))
    ]
    return filtered_od

# ê° íŒŒì¼ì— ëŒ€í•´ í•„í„°ë§ ì ìš©
filtered_od_list = []
for file in od_csv_files_filtered:
    filtered_od = filter_od_data(file)
    filtered_od_list.append(filtered_od)

# ëª¨ë“  í•„í„°ë§ëœ ë°ì´í„°ë¥¼ í•˜ë‚˜ì˜ DataFrameìœ¼ë¡œ ë³‘í•©
filtered_od_df = pd.concat(filtered_od_list, ignore_index=True)

print(filtered_od_df.head())
```

```python
# í¬í•¨í•  ëª©ì  ì½”ë“œ
include_purpose = [3, 5]  # ì‡¼í•‘/ì—¬ê°€, ì—¬í–‰

# Stay ë°ì´í„° íŒŒì¼ ê²½ë¡œ (ì••ì¶• í•´ì œëœ íŒŒì¼)
stay_csv_files = glob.glob('/content/data/stay/*.csv')

# ì›í•˜ëŠ” ë‚ ì§œ ë²”ìœ„ì˜ Stay íŒŒì¼ë§Œ ì„ íƒ
stay_csv_files_filtered = filter_files_by_date(stay_csv_files, start_date, end_date)

# Stay ë°ì´í„°ë¥¼ í•„í„°ë§í•˜ëŠ” í•¨ìˆ˜
def filter_stay_data(file):
    stay_df = pd.read_csv(file, dtype={'hdong_cd': 'int64', 'date': 'int64',
                                       'stay_cnts': 'int32', 'purpose': 'int8'})

    # ë‚ ì§œ í•„í„°ë§
    stay_df['date'] = pd.to_datetime(stay_df['date'], format='%Y%m%d')
    stay_df = stay_df[(stay_df['date'] >= start_date) & (stay_df['date'] <= end_date)]

    # í•„í„°ë§ ì¡°ê±´ ì ìš©
    filtered_stay = stay_df[
        (stay_df['hdong_cd'].isin(dest_codes)) &
        (stay_df['purpose'].isin(include_purpose))
    ]
    return filtered_stay

# ê° íŒŒì¼ì— ëŒ€í•´ í•„í„°ë§ ì ìš©
filtered_stay_list = []
for file in stay_csv_files_filtered:
    filtered_stay = filter_stay_data(file)
    filtered_stay_list.append(filtered_stay)

# ëª¨ë“  í•„í„°ë§ëœ ë°ì´í„°ë¥¼ í•˜ë‚˜ì˜ DataFrameìœ¼ë¡œ ë³‘í•©
filtered_stay_df = pd.concat(filtered_stay_list, ignore_index=True)
print(filtered_stay_df.head())
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

# 1. ë°ì´í„° ë¡œë“œ ë° ì „ì²˜ë¦¬ (ë°ì´í„°í”„ë ˆì„ì´ ì´ë¯¸ ì¡´ì¬í•œë‹¤ê³  ê°€ì •)
# od_df, stay_df, youtube_df

# ë‚ ì§œ í˜•ì‹ ë³€í™˜
od_df['date'] = pd.to_datetime(od_df['date']).dt.date
stay_df['date'] = pd.to_datetime(stay_df['date']).dt.date
youtube_df['Published_At'] = pd.to_datetime(youtube_df['Published_At']).dt.date

# ì‹œì¥ë³„ í‚¤ì›Œë“œ ì„¤ì •
yesan_keywords = ['ì˜ˆì‚°', 'ì˜ˆì‚°ì‹œì¥', 'ì˜ˆì‚° ë§¥ì£¼ í˜ìŠ¤í‹°ë²Œ']
seomun_keywords = ['ì„œë¬¸ì‹œì¥', 'ëŒ€êµ¬ ì¹˜ë§¥ í˜ìŠ¤í‹°ë²Œ']

# ì‹œì¥ë³„ ë°ì´í„° ë¶„ë¦¬
yesan_od = od_df[od_df['ì‹œì¥ëª…'] == 'ì˜ˆì‚°ì‹œì¥'].copy()
seomun_od = od_df[od_df['ì‹œì¥ëª…'] == 'ì„œë¬¸ì‹œì¥'].copy()

yesan_stay = stay_df[stay_df['ì‹œì¥ëª…'] == 'ì˜ˆì‚°ì‹œì¥'].copy()
seomun_stay = stay_df[stay_df['ì‹œì¥ëª…'] == 'ì„œë¬¸ì‹œì¥'].copy()

youtube_yesan = youtube_df[youtube_df['Search_Query'].isin(yesan_keywords)].copy()
youtube_seomun = youtube_df[youtube_df['Search_Query'].isin(seomun_keywords)].copy()

# ì¼ë³„ ë°©ë¬¸ì ìˆ˜ ê³„ì‚°
yesan_od_daily = yesan_od.groupby('date')['od_cnts'].sum().reset_index()
seomun_od_daily = seomun_od.groupby('date')['od_cnts'].sum().reset_index()

yesan_stay_daily = yesan_stay.groupby('date')['stay_cnts'].sum().reset_index()
seomun_stay_daily = seomun_stay.groupby('date')['stay_cnts'].sum().reset_index()

# ìœ íŠœë¸Œ ì¸ê¸° ì§€í‘œ ê³„ì‚°
def calculate_youtube_metrics(youtube_df):
    daily_metrics = youtube_df.groupby('Published_At').agg({
        'View_Count': 'sum',
        'Like_Count': 'sum',
        'Comment_Count': 'sum'
    }).reset_index().rename(columns={'Published_At': 'date'})
    daily_metrics['log_view_count'] = np.log1p(daily_metrics['View_Count'])
    daily_metrics['log_like_count'] = np.log1p(daily_metrics['Like_Count'])
    daily_metrics['log_comment_count'] = np.log1p(daily_metrics['Comment_Count'])
    return daily_metrics

youtube_yesan_daily = calculate_youtube_metrics(youtube_yesan)
youtube_seomun_daily = calculate_youtube_metrics(youtube_seomun)

# ë°ì´í„° ë³‘í•©
def merge_data(daily_df, youtube_daily_df):
    merged_df = daily_df.merge(
        youtube_daily_df[['date', 'log_view_count', 'log_like_count', 'log_comment_count']],
        on='date',
        how='left'
    )
    merged_df[['log_view_count', 'log_like_count', 'log_comment_count']] = merged_df[['log_view_count', 'log_like_count', 'log_comment_count']].fillna(0)
    return merged_df

yesan_od_daily = merge_data(yesan_od_daily, youtube_yesan_daily)
seomun_od_daily = merge_data(seomun_od_daily, youtube_seomun_daily)

yesan_stay_daily = merge_data(yesan_stay_daily, youtube_yesan_daily)
seomun_stay_daily = merge_data(seomun_stay_daily, youtube_seomun_daily)

# ì¶•ì œ ë° ì—°íœ´ ê¸°ê°„ ë³€ìˆ˜ ì¶”ê°€
festival_start_date = pd.to_datetime('2023-08-30').date()
festival_end_date = pd.to_datetime('2023-09-03').date()
holiday_start_date = pd.to_datetime('2023-09-28').date()
holiday_end_date = pd.to_datetime('2023-10-03').date()

def period_label(date):
    if festival_start_date <= date <= festival_end_date:
        return 'festival'
    elif holiday_start_date <= date <= holiday_end_date:
        return 'holiday'
    else:
        return 'normal'

for df in [yesan_od_daily, seomun_od_daily, yesan_stay_daily, seomun_stay_daily]:
    df['period'] = df['date'].apply(period_label)

# ì²˜ë¦¬ ê·¸ë£¹ ë° í†µì œ ê·¸ë£¹ êµ¬ë¶„
yesan_od_daily['group'] = 1
seomun_od_daily['group'] = 0
yesan_stay_daily['group'] = 1
seomun_stay_daily['group'] = 0

# ë°ì´í„° ë³‘í•© ë° ì •ë ¬
did_od_data = pd.concat([yesan_od_daily, seomun_od_daily], ignore_index=True).sort_values(by=['group', 'date']).reset_index(drop=True)
did_stay_data = pd.concat([yesan_stay_daily, seomun_stay_daily], ignore_index=True).sort_values(by=['group', 'date']).reset_index(drop=True)

# ì‹œì°¨ ë³€ìˆ˜ ìƒì„±
def create_lagged_variables(did_data, lag=1):
    for col in ['log_view_count', 'log_like_count', 'log_comment_count']:
        did_data[f'{col}_lag{lag}'] = did_data.groupby('group', sort=False)[col].shift(lag).fillna(0)
    return did_data

# 7ì¼ ì‹œì°¨ ë³€ìˆ˜ ìƒì„±
did_od_data = create_lagged_variables(did_od_data, lag=7)
did_stay_data = create_lagged_variables(did_stay_data, lag=7)

# ìƒí˜¸ì‘ìš© í•­ ìƒì„±
def create_interaction_terms(did_data):
    for col in ['log_view_count_lag1', 'log_like_count_lag1', 'log_comment_count_lag1']:
        did_data[f'group_{col}'] = did_data['group'] * did_data[col]
    return did_data

did_od_data = create_interaction_terms(did_od_data)
did_stay_data = create_interaction_terms(did_stay_data)

# ê¸°ê°„ ë”ë¯¸ ë³€ìˆ˜ ìƒì„± ë° ìƒí˜¸ì‘ìš© í•­ ìƒì„± (ìˆ˜ì •ëœ í•¨ìˆ˜ ì‚¬ìš©)
def create_period_dummies(did_data):
    # drop_first ì˜µì…˜ ì œê±°í•˜ì—¬ ëª¨ë“  ë”ë¯¸ ë³€ìˆ˜ê°€ ìƒì„±ë˜ë„ë¡ í•¨
    did_data = pd.get_dummies(did_data, columns=['period'])
    for col in ['period_festival', 'period_holiday']:
        if col in did_data.columns:
            did_data[f'group_{col}'] = did_data['group'] * did_data[col]
        else:
            # í•´ë‹¹ ê¸°ê°„ì— ë°ì´í„°ê°€ ì—†ìœ¼ë©´ ì»¬ëŸ¼ì„ ìƒì„±í•˜ê³  0ìœ¼ë¡œ ì±„ì›€
            did_data[col] = 0
            did_data[f'group_{col}'] = 0
    return did_data

did_od_data = create_period_dummies(did_od_data)
did_stay_data = create_period_dummies(did_stay_data)

# íšŒê·€ ëª¨ë¸ ì •ì˜ (ëª¨ë¸ ê³µì‹ ìˆ˜ì •)
od_formula_terms = ['group']
od_formula_terms += [col for col in ['period_festival', 'period_holiday'] if col in did_od_data.columns]
od_formula_terms += ['log_view_count_lag1', 'log_like_count_lag1', 'log_comment_count_lag1']
od_formula_terms += [col for col in ['group_period_festival', 'group_period_holiday'] if col in did_od_data.columns]
od_formula_terms += ['group_log_view_count_lag1', 'group_log_like_count_lag1', 'group_log_comment_count_lag1']

model_formula_od = 'od_cnts ~ ' + ' + '.join(od_formula_terms)

stay_formula_terms = ['group']
stay_formula_terms += [col for col in ['period_festival', 'period_holiday'] if col in did_stay_data.columns]
stay_formula_terms += ['log_view_count_lag1', 'log_like_count_lag1', 'log_comment_count_lag1']
stay_formula_terms += [col for col in ['group_period_festival', 'group_period_holiday'] if col in did_stay_data.columns]
stay_formula_terms += ['group_log_view_count_lag1', 'group_log_like_count_lag1', 'group_log_comment_count_lag1']

model_formula_stay = 'stay_cnts ~ ' + ' + '.join(stay_formula_terms)

# íšŒê·€ ë¶„ì„ ìˆ˜í–‰
model_did_od = smf.ols(model_formula_od, data=did_od_data).fit(cov_type='HC1')
model_did_stay = smf.ols(model_formula_stay, data=did_stay_data).fit(cov_type='HC1')

# ê²°ê³¼ ì¶œë ¥
print("DID ë¶„ì„ ê²°ê³¼ (OD ë°ì´í„°)")
print(model_did_od.summary())

print("\nDID ë¶„ì„ ê²°ê³¼ (Stay ë°ì´í„°)")
print(model_did_stay.summary())
```
