```python
import pandas as pd
import json
```

```python
#좌표 데이터 불러오기
with open('network_YG_edge_youtube.json', 'r') as f:
    data = json.load(f)

coordinates = pd.DataFrame([{
    'id': node['attributes']['label'],
    'x': node['attributes']['x'],
    'y': node['attributes']['y'],
} for node in data['nodes']])

# 링크 데이터 불러오기
links = pd.read_csv('network_YG_edge_youtube.csv')
# 모든 열 이름의 앞뒤 공백 제거 및 소문자로 변환
links.columns = links.columns.str.strip().str.lower()
```

```python
links.drop(columns='id', inplace = True)
```

```python
coordinates
```

```python
#링크 데이터에 relationship 컬럼 생성
links['relationship'] = [pair[0] + '->' + pair[1] for pair in zip(links['source'], links['target'])]
```

```python
# 태블로형 데이터로 변환
links_re = links[['target', 'weight', 'relationship']] #target ~ relationship 컬럼을 복사한 데이터프레임
links_re = links_re.rename(columns={'target': 'id'}) #두 데이터프레임을 합치기 위해 컬럼 이름 변경
links = links.rename(columns={'source': 'id'}) #두 데이터프레임을 합치기 위해 컬럼 이름 변경
del links['target'] #필요없는 target 컬럼 제거
links_concated = pd.concat([links, links_re]) #데이터 합치기
```

```python
links_concated
```

```python
#좌표값을 merge해줍니다.
links_concated = pd.merge(links_concated, coordinates, on='id', how='left')

links_concated.to_csv('YG_youtube_links_concated.csv') #데이터 저장
```
